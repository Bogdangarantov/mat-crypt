```python
import sys
import time
import math
import random
import numpy as np
from sympy import primerange
from KupynaRealization import KupynaRealization
from ByteIter import ByteIter
random.seed(48792357293)
```

```python
def modify_single_bit(b):
    b[random.randint(0, len(b) - 1)] ^= 1 << random.randint(0, 7)
    return b
def hash_difference(h1, h2):
    return sum((x ^ y).bit_count() for x, y in zip(h1, h2))
def is_partial_collision(h1, h2, level=0.3):
    diff = hash_difference(h1, h2)
    l = len(h1) * 8
    return diff / l > level
```

```python
K = KupynaRealization(512)
table = [[float("inf") for _ in range(10)] for __ in range(5, 16)]

level = 0.52

for j, k in enumerate(range(5, 16)):
    for i in range(10):
        t = time.time()
        b1 = None
        for b2 in ByteIter(k):
            if b1 is None:
                b1 = b2
                continue
            h1, h2 = K.messageHashing(b1), K.messageHashing(b2)
            if is_partial_collision(h1, h2, level):
                t = time.time() - t
                table[j][i] = t
                break

for i, x in enumerate(table):
    print(f"Для k = {(i + 5)}: ", x)
    print(f"Середнє арифметичне: {np.mean(x)}")
    print(f"Відхилення: {np.std(x)}")
```
